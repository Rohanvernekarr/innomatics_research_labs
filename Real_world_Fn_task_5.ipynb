Social Media – Post Engagement Analyzer

```python
def analyze_engagement(likes):
    total = 0

    for like in likes:
        total += like

    if total >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total)
    print("Post Status:", status)

likes = [200, 300, 150, 400]

analyze_engagement(likes)
```

Agriculture – Rainfall Adequacy Checker

```python
def check_rainfall(rainfall, required_level):
    total = 0

    for value in rainfall:
        total += value

    average = total / len(rainfall)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", int(average))
    print("Rainfall Status:", status)

rainfall = [56, 65, 72, 78, 89]
required_level = 60

check_rainfall(rainfall, required_level)
```

Social Media – Duplicate Account Detection

```python
def detect_duplicate_accounts(usernames):
    unique = set(usernames)

    if len(unique) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

usernames = ["user1", "user2", "user2", "user4"]

detect_duplicate_accounts(usernames)
```

 Agriculture – Premium Crop Price Filter

```python
def premium_crops(prices):
    result = []
    for price in prices:
        if price > 2000:
            result.append(price)
    print("Premium Crops:", result)


# user nput
n = int(input("Enter no of crops: "))
prices = []
for i in range(n):
    prices.append(int(input(f"Enter the price for crop {i+1}: ")))

# function call
premium_crops(prices)
```

System Monitoring – Application Health Checker

```python
def check_app_health(errors):
    # system status
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    # Print results
    print("Error Count:", errors)
    print("System Status:", status)

errors = int(input("Enter number of errors: "))

check_app_health(errors)
```

 Daily Transaction Limit Checker

```python
def check_transaction(amount):
    daily_limit = 50000
    print(f"Transaction Amount: {amount}")

    if amount <= daily_limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")

amount = float(input("Enter transaction amount: ₹"))

check_transaction(amount)
```

 Student Attendance Eligibility System

```python
def check_exam_eligibility(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

attendance = ["P", "A", "A", "P", "P"]

check_exam_eligibility(attendance)
```

Smart Electricity Bill Analyzer

```python
def calculate_bill(units):

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = 100 * 3 + (units - 100) * 5
    else:
        bill = 100 * 3 + 100 * 5 + (units - 200) * 7


    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage

units = float(input("Enter electricity units consumed: "))

bill, usage_status = calculate_bill(units)
print(f"Total Bill: ₹{bill}")
print(f"Usage Classification: {usage_status}")
```
