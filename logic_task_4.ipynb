Mobile Recharge Validation System



```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be at least ₹50")
        return False

    if amount not in valid_plans:
        print("Invalid plan. Available plans are:", valid_plans)
        return False

    print("Recharge successful for ₹", amount)
    return True
#retry
while True:
    amount = int(input("Enter recharge amount: "))

    if validate_recharge(amount):
        break   # exit loop if recharge is valid
    else:
        print("Please try again.\n")
```

 Inventory Reorder Alert System

```python
def check_inventory(products):

    for item in products:
        quantity = products[item]

        if quantity < 15:
            print(item, ": Reorder Alert")
        else:
            print(item, ": Stock OK")

products = {
    "Rice": 10,
    "Sugar": 25,
    "Oil": 12,
    "Salt": 30
}

check_inventory(products)
```

Student Result Processing System


```python
def process_result(marks):

    total = 0
    for mark in marks:
        total = total + mark

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        print("Result: Pass")
    else:
        print("Result: Fail")

student_marks = [60, 55, 45, 70]

process_result(student_marks)
```

Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, peak):
    base_fare = 50
    perkmcharge = 12

    fare = base_fare + (distance * perkmcharge)
    if peak == "yes":
        fare = fare + (fare * 0.25)

    return fare
# Retry
while True:
    distance = int(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()

    total_fare = calculate_fare(distance, peak)

    print("Total Cab Fare: Rs", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

 Employee Attendance Eligibility Checker


```python
def check_attendance(attendance):
    present_count = 0

    for day in attendance:
        if day == "P":
            present_count = present_count + 1

    percentage = (present_count / len(attendance)) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


#list
attendance_list = ["P", "P", "P", "P", "P", "A"]

# Call the function
result = check_attendance(attendance_list)
print("Status:", result)

```

Password Strength Checker

```python
def check_password(password):

    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return "Weak Password (Minimum length is 8)"

    for ch in password:
        if ch.isdigit():
            has_digit = True

        if ch in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain digit and special character)"

user_password = input("Enter your password: ")

result = check_password(user_password)
print(result)
```
