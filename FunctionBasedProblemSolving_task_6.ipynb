Smart Parking Lot Management System


```python
def smart_parking_system(capacity, vehicle_logs):
    parked = 0
    peak = 0

    for log in vehicle_logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        if parked > peak:
            peak = parked

    print("Currently Parked Vehicle:", parked)

    if parked > capacity:
        print("Parking Status Full Over Capacity Alert!")
    else:
        print("Parking Status Available")

    print("Peak Usage:", peak)

smart_parking_system(50, ["IN", "IN", "OUT", "OUT", "IN", "IN", "OUT"])
```

Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5

    # Traffic
    if traffic.lower() == "high":
        base_time += 15
    elif traffic.lower() == "medium":
        base_time += 10

    # Weather
    if weather.lower() == "rainy":
        base_time += 10
    elif weather.lower() == "storm":
        base_time += 20

    print("Estimated Delivery Time:", base_time, "minutes")

estimate_delivery_time(8, "High", "Rainy")
```

Movie Theatre Seat Occupancy Analyzer

```python
def theatre_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", round(occupancy, 2), "%")

    if occupancy == 100:
        print("Show Status Housefull")
    elif occupancy >= 75:
        print("Show Status Almost Full")
    else:
        print("Show Status Available")

theatre_occupancy(200, [1]*150)
```

Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    average = sum(cpu_readings) / len(cpu_readings)

    print("Average CPU Load:", round(average, 2), "%")

    if average < 50:
        print("Server Normal")
    elif 50 <= average <= 80:
        print("Server Warning")
    else:
        print("Server Critical")

classify_server_load([45, 55, 70, 80, 90])
```

Smart Classroom Resource Usage Monitor

```python
def classroom_monitor(resource_usage):
    overused = []

    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert No")

classroom_monitor({
    "Projector": 6,
    "AC": 10,
    "Lights": 4
})
```

Online Event Registration Capacity Controller

```python
def event_registration(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)

event_registration(180, 185)
```
