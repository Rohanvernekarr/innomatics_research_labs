Smart Light Controller

```python
class SmartLight:
    def __init__(self,light_name):
        self.light_name = light_name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.light_name} is {self.status}")

light_name = input("Light Name: ")
action = input("Action On or off: ").upper()

light = SmartLight(light_name)

if action == "ON":
    light.turn_on()
elif action == "OFF":
    light.turn_off()
else:
    print("Invalid Action")

light.display_status()
```

 Employee ID Card System


```python
class Employee:
    def __init__(self,name,emp_id,department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print("Name :",self.name)
        print("ID :",self.emp_id)
        print("Department :",self.department)

name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")

emp = Employee(name, emp_id, department)
emp.display_id_card()
```

Mobile Contact Record


```python
class Contact:
    def __init__(self,name,phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print("Name:",self.name)
        print("Phone:",self.phone)

name = input("Contact Name: ")
phone = input("Phone Number: ")

contact = Contact(name, phone)
contact.display_contact()
```

Product Price Tag Generator


```python
class Product:
    def __init__(self,name,price):
        self.name = name
        self.price = price

    def print_price_tag(self):
        print("Product:",self.name)
        print("Price: Rs"+str(self.price))

name = input("Product Name: ")
price = input("Price: ")

product = Product(name, price)
product.print_price_tag()
```

Delivery Address Manager


```python
class Customer:
    def __init__(self,name,address):
        self.name = name
        self.address = address

    def print_delivery_details(self):
        print("Delivery Details")
        print("Customer:",self.name)
        print("Address:",self.address)

name = input("Customer Name: ")
address = input("Address: ")

customer = Customer(name, address)
customer.print_delivery_details()
```

 Movie Rating Display System


```python
class Movie:
    def __init__(self,name,rating):
        self.name = name
        self.rating = rating

    def display_movie(self):
        print("Movie:",self.name)
        print("Rating:",self.rating, "/ 5")

name = input("Movie Name: ")
rating = input("Rating: ")

movie = Movie(name, rating)
movie.display_movie()
```
