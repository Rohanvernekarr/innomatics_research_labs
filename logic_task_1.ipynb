```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

# for loop to iterate
for m in marks:
    if m >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Pass Students:", pass_count)
print("Total Fail Students:", fail_count)
```

```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

# use append

for name in names:
    cleaned_names.append(name.strip().lower())

print("Cleaned Names:", cleaned_names)
```

```python

messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    print(f"Message: '{msg}' | Length: {length}")

    if length > 10:
        print("Long message detected")
```
